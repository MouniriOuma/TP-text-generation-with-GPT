```python
from transformers import pipeline
from transformers import set_seed
import os
import random
```

```python
def generate_story(initial_phrase, model_name="gpt2", max_length=200, num_return_sequences=1, seed=None, temperature=1.0, top_p=0.9, repetition_penalty=1.2):
    if seed is not None:
        set_seed(seed)

    generator = pipeline('text-generation', model=model_name)

    generated_text = generator(
        initial_phrase,
        max_length=max_length,
        num_return_sequences=num_return_sequences,
        temperature=temperature,
        top_p=top_p,
        repetition_penalty=repetition_penalty,
    )

    return [seq['generated_text'] for seq in generated_text]
```

```python
 # 1. Call generate_story with initial phrase
text1 = generate_story(initial_phrase="My idea from the start was ...")
print("Text 1:\n")
for text in text1:
        print(text)
```

```python
# 2. Call generate_story with initial phrase, seed, temperature, top_p, max_length
text2 = generate_story(
    initial_phrase="My idea from the start was ...",
    seed=42,
    temperature=1.2,
    top_p=0.95,
    max_length=300
)

print("Text 2:\n")

for text in text2:
        print(text)
```

```python
# 3. Call generate_story 3 times with random seed and max_length=150
for _ in range(3):
    random_seed = random.randint(0, 10000)
    text3 = generate_story(
        initial_phrase="My idea from the start was ...",
        seed=random_seed,
        max_length=150
    )
print(f"Text 3 (seed: {random_seed}):\n")

for text in text3:
  print(text)
```

```python
# 4. Call generate_story with initial phrase and different model
text4 = generate_story(
    initial_phrase="My idea from the start was ...",
    model_name="gpt2-medium",
    max_length=250
)
print("Text 4:\n")

for text in text4:
  print(text)
```
